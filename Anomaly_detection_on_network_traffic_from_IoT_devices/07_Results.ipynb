> The accuracy for all the classifiers, other than the KNeighbors, RandomForest, SVC with rbf kernel, is about 1% lower than the rest of the other metrics. This can be explained by the fact that the True Negatives (negative being the benign class) are less than the True Positives (positive being the malicious classes) as we have seen across all tests. The proof of this can be derived from the following for example: 

<br>
<br>



$$
  \mathbf{Accuracy} < \mathbf{F1} \Leftrightarrow
$$

$$
  \frac{TP+TN}{TP+TN+FP+FN} < \frac{2TP}{2TP+FP+FN} \Leftrightarrow
$$

$$
  (TP+TN)\cdot(2TP+FP+FN) < 2TP \cdot(TP+TN+FP+FN) \Leftrightarrow 
$$

$$
  \cancel{2TP^2} + TP\cdot FP + TP\cdot FN + \cancel{2TP\cdot TN} + TN\cdot FP + FN\cdot TN < \cancel{2TP^2} + \cancel{2TP\cdot TN} + 2TP\cdot FP + 2TP\cdot FN \Leftrightarrow
$$

$$
  TN\cdot FP + FN\cdot TN < TP\cdot FP + TP\cdot FN \Leftrightarrow 
$$

$$
  TN\cdot ~~(FP+FN)~~ < TP\cdot ~~(FP+FN)~~ \Leftrightarrow
$$

$$
  \mathbf{TN} < \mathbf{TP}
$$
